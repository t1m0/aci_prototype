```python
from Application.PaparaziGame import PaparaziGame
from Application.plot_map import plot_map
import pandas as pd
```

```python
game = PaparaziGame()
```

```python
game_map = game.play("Maps/",num_security_guards=10, iterations=10)
```

```python
columns=['name', 'time_elapsed', 'total_iteration', 'total_cost', 'path_length', 'heuristic', 'map','path']
df = pd.DataFrame([[getattr(i,j) for j in columns] for i in game_map], columns=columns)
```

```python
df
```

Show performance of best iterations

```python
df.groupby(['map','heuristic'])[['total_cost', 'time_elapsed']].min()
```

Show mean performance of the heuristics

```python
df.groupby(['map','heuristic'])[['total_cost', 'time_elapsed']].mean()
```

Show median performance of the heuristics

```python
df.groupby(['map','heuristic'])[['total_cost', 'time_elapsed']].median()
```

* For the map size 100 we can see no signifcant gap between mean and median -> normal distribution
* For map size 25 and 50 we see a significatn gap between mean and median -> no normal distribution
* Due to the smaller map size the random placement of the security guards has a significant impact.
* For larger maps Chebyshev seems to produce the lowest cost (even though we have slightly slower exection time)
